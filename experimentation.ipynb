```python
train_data, test_data = train_test_split(churn_data,test_size = 0.30, random_state =42)
```

```python
train_data.shape, test_data.shape
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
label = 'churn'
```

```python
print("Summary of churn variable:\n", train_data[label].describe())
```

## basic model build

```python
%%time

save_path = 'agModels-predictchurn'
predictor = TabularPredictor(label=label, path=save_path).fit(train_data)
```

```python
predictor.leaderboard(silent=True)
```

## Improving Predictive Performance

Our basic model build took 31.9 seconds, we can improve our model by setting a shorter time limit of 25 seconds

```python
#time_limit = 25
```

We can also set an evaluation metric

```python
#metric = 'roc_auc'
```

```python
#save_path = 'agModels-predictchurnimproved'
```

.. and also preset our models for best quality fit

```python
%%time

time_limit = 20
save_path = 'AutogluonModels'
metric = 'roc_auc'
predictor = TabularPredictor(label, eval_metric=metric).fit(train_data, time_limit=time_limit, presets='best_quality')
```

```python
predictor.leaderboard(test_data, silent=True)
```
